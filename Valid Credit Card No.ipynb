```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('cc.txt',header=None, names=['cc'])
```

```python
df
```

```python
def valid(x):
    if re.findall('[^\d-]',x):
        return 'Invalid'
    
    if len(re.sub('\D','',x)) != 16:
        return 'Invalid'
    
    matched = re.match('^([456]\d{3})-?(\d{4})-?(\d{4})-?(\d{4})$',x)
    if matched:
        for i in range(4):
            b = re.sub('\D','',x)[i:]
            #print(b)
            pattern = "^(\d{4})(\d{4})(\d{4})(\d{%d})$" % (4-i)
            #print(pattern)
            matched_2 = re.match(pattern,b)
            #print(matched_2.groups())
            for group in matched_2.groups():
                if group[0]*4 == group:
                    return 'Invalid'
    else:
        return 'Invalid'
    
    return 'Valid'
    
```

```python
df['Result'] = df['cc'].apply(valid)
```

```python
def valid_2(x):
    if re.findall('[^\d-]',x):
        return 'Invalid'
    
    if len(re.sub('\D','',x)) != 16:
        return 'Invalid'
    
    if re.match('^([456]\d{3})-?(\d{4})-?(\d{4})-?(\d{4})$',x):
        b = re.sub('\D','',x)
        num = 0
        l = []
        for i in b:
            l.append(i)
            #print(l)
            if num >= 4:
                l = l[1:]
                #print(l)
            num += 1
            
            if list(l[0]*4) == l:
                return 'Invalid'
    else:
        return 'Invalid'
    
    return 'Valid'
    
```

```python
df['Result_2'] = df['cc'].apply(valid_2)
df
```
